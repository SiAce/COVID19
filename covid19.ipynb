```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from matplotlib.pyplot import figure
```

```python
pd.set_option('display.max_rows', None)
```

```python
covid_country = pd.read_csv("country.csv")
```

```python
covid_usa = covid_country[covid_country['countryterritoryCode'] == 'USA'].sort_index(ascending=False).reset_index(drop=True)
```

```python
date = covid_usa['dateRep']
daily_cases = covid_usa['cases'].to_numpy()
total_cases = daily_cases.cumsum()
```

```python
growth_rate = daily_cases / total_cases
```

```python
figure(num=None, figsize=(20, 15), dpi=100, facecolor='w', edgecolor='k')
plt.xticks(rotation=90)
plt.bar(date, growth_rate)
```

```python
plt.bar(date, daily_cases)
```

```python
plt.bar(date, total_cases)
```
